###  Selecione a primeira transação diária de cada cliente

```python

transacoes = pd.read_csv ("../data/transacoes.csv", sep=";")
transacoes
```

```python
transacoes = transacoes.sort_values("DtCriacao") #ordena
transacoes["data"] = pd.to_datetime(transacoes["DtCriacao"]).dt.date #cria coluna apenas da data
transacoes.drop_duplicates(keep="first", subset=["IdCliente", "data"]) #dropa as mesmas datas e mantem a primeira
```
