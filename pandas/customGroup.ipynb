```python
import pandas as pd
import numpy as np
```

```python
df_transacoes = pd.read_csv ("data/transacoes.csv", sep=";")
df_transacoes
```

```python
def diff_amp (x: pd.Series):
    amplitude = x.max() - x.min()
    media = x.mean()
    return np.sqrt((amplitude-media)**2)
```

```python
(df_transacoes.groupby(by=["IdCliente"]).agg({"IdTransacao":['count'],
                                              "QtdePontos": ["sum","mean",diff_amp]})
                                              )
```
