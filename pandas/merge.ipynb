```python
import pandas as pd

df_transacoes = pd.read_csv ("data/transacoes.csv", sep= ";")
df_transacoes.head()
```

```python
df_clientes = pd.read_csv ("data/clientes.csv", sep=";")
df_clientes.head()
```

```python
df_transacoes.merge(right=df_clientes, 
                    how='left', 
                    left_on=["IdCliente"], 
                    right_on=["idCliente"],
                    suffixes=["Transacao","Cliente"]) #para colunas repetidas do mesmo dataset, usa sufixo
```

```python
df_transacao_produto = pd.read_csv("data/transacao_produto.csv", sep=";")
df_transacao_produto
```

```python
df_produto = pd.read_csv("data/produtos.csv", sep=";")
df_produto.head(14)
```

```python
cliente_transacao_produto = df_transacoes.merge(df_transacao_produto, on="IdTransacao",how="left")
```

```python
#filtrando
cliente_transacao_produto[["IdTransacao","IdCliente","IdProduto"]]
```

```python
df_full = cliente_transacao_produto.merge(df_produto, on=["IdProduto"], how="left")
df_full.head()
```

```python
df_full = df_full[df_full["DescNomeProduto"]=="Presença Streak"]
df_full.groupby(by=["IdCliente"])["IdTransacao"].count().sort_values(ascending=False).head(1)
```
