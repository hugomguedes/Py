```python
import pandas as pd
import os

df_geral = pd.read_csv("data/ipea/homicidios.csv", sep=";")
df_geral.head()
```

```python
# df_geral = df_geral.rename(columns={"valor": "homicidios"})
# df_geral = df_geral.drop(["cod"], axis=1)
# df_geral.head()
def read_file (file_name:str):
    df = (pd.read_csv(f"data/ipea/{file_name}.csv",sep=";").rename(columns={"valor":file_name}).set_index(["nome","período"]).drop(["cod"],axis=1))
    return df
```

```python
df_negros = read_file("homicidios-negros")
df_negros
```

```python
files_names = os.listdir("data/ipea/")
files_names
```

```python
dfs = []
for i in files_names:
    file_name = i.split(".")[0]
    dfs.append(read_file(file_name))
```

```python
dfs[-3]
```

```python
df_full =pd.concat(dfs, axis=1).reset_index().sort_values(["período", "nome"])
df_full
```

```python
df_full.to_csv("homicidios_consolidado.csv",index=False, sep= ";")
```
