```python
import pandas as pd

df = pd.read_csv("homicidios_consolidado.csv", sep=";")
df.head()
```

```python
df_stack = (df.set_index(["nome","período"]).stack().reset_index())

df_stack.columns = ["nome", "periodo", "metrica", "valor"]
df_stack
```

```python
df_stack.pivot_table(values="valor",index=["nome","periodo"],columns="metrica")
```

```python
(df_stack.pivot_table(values="valor",index=["nome","periodo"],columns="metrica").reset_index())
```

```python
df_stack.pivot_table(values="valor",index=["nome"],columns="metrica",aggfunc="mean")
```
