```python
import pandas as pd

df = pd.read_csv("homicidios_consolidado.csv", sep=";")
df.head()
```

```python
df = df.reset_index()
# df =df.set_index (["nome", "período"])
# df_stack =df.stack()
# df_stack.head()
```

```python
df = df.set_index(["nome", "período"])
df_stack = df.stack()
df_stack.head()
```

```python
df_stack = df_stack.reset_index()
```

```python
df_stack.columns = ["nome", "período", "metrica", "valor"]
df_stack.head()
```

```python
df_unstack = (df_stack.set_index(["nome", "período", "metrica"]).unstack().reset_index())
df_unstack
```

```python
metricas = df_unstack.columns.droplevel(0)[2:].to_list() #excluindo multi-index
df_unstack.columns = ["nome", "período"] + metricas
df_unstack
```
