```python
import pandas as pd
```

```python
df = pd.DataFrame ({
    "cliente": [1,2,3,4,5],
    "nome": ["teo","jose","maria","alfredo","romeu"]
})

df_02 = pd.DataFrame ({
    "cliente": [6,7,8],
    "nome": ["afonso","shirley","pedro"],
    "idade": [32,29,31]
})

df_03 = pd.DataFrame ({
    "idade": [32,34,19,54,33]
})
dfs = [df,df_02]
```

```python
#concatena listas um em cima do outro
pd.concat(dfs, ignore_index=True)
```

```python
#concatena um ao lado do outro
pd.concat([df,df_03], axis=1) 
```

```python
#ordenou
df_03 = df_03.sort_values(by="idade")
df_03
```

```python
#está ordenando por index e não por valor de idade
pd.concat([df, df_03], axis=1)
```

```python
df_03 = df_03.sort_values(by="idade").reset_index(drop=True)
df_03
```

```python
pd.concat([df, df_03], axis=1)
```
